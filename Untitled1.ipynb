тут удаляется код, теги, все символы кроме английских слов (другие языки тоже) 
очищаются теги

```
from sklearn.metrics import  mean_squared_error
import re
from nltk.corpus import stopwords

#remove code so it won't affect results
def remove_code(text_to_clean):
  return re.sub(r'<code>(?s).*</code>', '', text_to_clean)

#remove html tags
def remove_html(text_to_clean):
  return re.sub(r'<.*?>', ' ', text_to_clean)

#remove everything but words
def remove_simbols(text_to_clean):
  text_to_clean = re.sub('[^a-zA-Z]',' ', text_to_clean)
  return re.sub(' +', ' ', text_to_clean)

#clean "<" and ">" from tags
def clean_tags(text_to_clean):
  return text_to_clean.replace('<', '').replace('>',' ').strip()


df_train.Body = df_train.Body.apply(remove_code)
df_train.Body = df_train.Body.apply(remove_html)
df_train.Body = df_train.Body.apply(remove_simbols)
df_train.Body = df_train.Body.str.lower().str.strip()

df_train.Tags = df_train.Tags.apply(clean_tags)
df_train.Tags = df_train.Tags.apply(remove_simbols)
df_train.Tags = df_train.Tags.str.lower().str.strip()

df_train.Title = df_train.Title.apply(remove_simbols)
df_train.Title = df_train.Title.str.lower().str.strip()

df_train.head()
```

тут загрузка всяких штук отдельно, чтобы лишний раз не загружить их

```
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
```

удаляем лишние слова (я. ты. могу, не могу, и проч)

```
train.head()
def remove_stopwords(text):
  stop_words = set(stopwords.words('english'))
  lemmatizer = WordNetLemmatizer()
  word_token = word_tokenize(text)
  filtered = ' '.join([lemmatizer.lemmatize(a) for a in word_token if a not in stop_words])
  return filtered
train['body'] = train['body'].apply(remove_stopwords)
```

```
from collections import Counter
import plotly.express as px
def common_words(df, class_type):
    df['temp_list'] = df['body'].apply(lambda x:str(x).split())

    top = Counter([item for sublist in df['temp_list'].loc[df['target'] == class_type] for item in sublist])
    top_class = pd.DataFrame(top.most_common(15))
    top_class.columns = ['words','count_words']

    fig = px.bar(top_class, x='count_words',y='words',title=f'Common words in class {class_type}',orientation='h',width=800,height=700,color='words')
    fig.show()
```

```
import gensim
from gensim.models import Word2Vec

word2vec_pretrain = gensim.models.KeyedVectors.load_word2vec_format('/content/drive/MyDrive/NLP/GoogleNews-vectors-negative300.bin.gz', binary= True)
```

```
print(word2vec_pretrain)
```

```
def get_model_vectors(corpus, model):    
    
    def get_vector(sentence):   
        word_tokens = word_tokenize(" ".join(sentence))
        vector = 0
        for w in word_tokens:
            try: 
                vector += model.wv[w]
            except:
                continue
        return vector
    
    out_corpus = np.stack(corpus.apply(lambda x: get_vector(x)).to_numpy())
    
    return out_corpus
```

```
sentences = [row.split() for row in train.body]    
vocab = set( [item for sublist in sentences for item in sublist]) 
```

```
word2vec_model = Word2Vec(size = 32, min_count = 5, window = 5)
word2vec_model.build_vocab(sentences=sentences)
```

```
word2vec_model.train(sentences=sentences, total_examples=len(sentences), epochs=10)
```

```
word2vec_vectors = get_model_vectors(train.body, word2vec_model)
```

```
word2vec_vectors_pretrained = get_model_vectors(train.body, word2vec_pretrain)
```
